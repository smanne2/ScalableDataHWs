```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import functools
import math
from collections import OrderedDict
```

```python
df = pd.read_csv("citibike.csv")
df
```

Problem (2.a) = how many trips were from gender 1, and how many trips were from gender 2. 

```python
rows = list(df.gender.values)
pairs = map(lambda x:(int(x==1),int(x==2)),rows)
pairs =  list(pairs)
count = functools.reduce(lambda x,y: (x[0]+y[0],x[1]+y[1]), pairs)
print (count)
```

Problem(2b) = count the number of trips per birth year using higher order functions

```python
by = list(df.birth_year)
by = [x for x in by if not (math.isnan(x))]
# print(years)
```

```python
def reducer(x,y):
    x[y] = x.get(y,0)+1
    return x

counts = functools.reduce(reducer, by, {})
counts = dict(OrderedDict(sorted(counts.items())))
print(counts)
```
